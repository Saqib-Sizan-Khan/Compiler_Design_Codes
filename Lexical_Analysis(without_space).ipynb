```python
syntex = 'int a = a +[{(b*c)+10+20}-d22]; // this is a comment'
syntex2 = list()

error = False
index = -1

kw = ['int', 'float', 'double', 'if', 'else if', 'else', 'for']
ao = ['+', '-', '*', '/', '=']
punct = [':', ';', ',']
pt = ['(', ')', '{', '}', '[', ']']

keyword = []
identifier = []
constant = []
arithmetic_operator = []
punctuation = []
parenthesis = []
bracket = []

print(f'Code Syntex: {syntex}',end='\n\n')

index = syntex.find(' //')
if index >= 0:
  syntex = syntex[0:index]

if syntex[-1] != ';':
  error = True

for t in syntex:
  if t in ao:
    syntex2.append(' ')
    syntex2.append(t)
    syntex2.append(' ')
  elif t in punct:
    syntex2.append(' ')
    syntex2.append(t)
    syntex2.append(' ')
  elif t in pt:
    syntex2.append(' ')
    syntex2.append(t)
    syntex2.append(' ')
  else:
    syntex2.append(t)

syntex2 = ''.join(syntex2)
tokens = syntex2.split()

for token in tokens:
  if token in kw:
    keyword.append(token)
  elif token in ao:
    arithmetic_operator.append(token)
  elif token.isdecimal():
    constant.append(token)
  elif token in pt:
    parenthesis.append(token)
  elif token in punct:
    punctuation.append(token)
  else:
    identifier.append(token)

print(f'Keyword ({len(set(keyword))}):', end=' ')
print(*keyword, sep=', ')

for i in identifier:
  if i[0].isdigit():
    error = True

print(f'Indentifier ({len(set(identifier))}):', end=' ')
print(*identifier, sep=', ')

print(f'Arithmetic Operator ({len(set(arithmetic_operator))}):', end=' ')
print(*arithmetic_operator, sep=', ')

print(f'Constant ({len(set(constant))}):', end=' ')
print(*constant, sep=', ')

for p in parenthesis:
  if p == '(':
    if ')' in parenthesis:
      bracket.append('()')
      parenthesis.remove(')')
    else:
      error = True
      bracket.append(p)

  if p == '{':
    if '}' in parenthesis:
      bracket.append('{}')
      parenthesis.remove('}')
    else:
      error = True
      bracket.append(p)

  if p == '[':
    if ']' in parenthesis:
      bracket.append('[]')
      parenthesis.remove(']')
    else:
      error = True
      bracket.append(p)
  
  elif p == ')' or p == '}' or p == ']':
    error = True
    bracket.append(p)

print(f'Parenthesis ({len(set(bracket))}):', end=' ')
print(*bracket, sep=', ')

print(f'Punctuation ({len(set(punctuation))}):', end=' ')
print(*punctuation, sep=', ')

if error:
  print("Appered Lexical Error")
else:
  print('No Lexical Error')
```
